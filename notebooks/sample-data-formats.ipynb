# Sample data formats

```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

## Images and Masks

For a segmentation task, you must have a set of (1) input images, and (2) associated masks.  The masks serve as the labels, and you need image/mask pairs to train your DNN.  Below is an example of an image and the associated mask (labels). 

The mask is shown in both .png and .npy file formats.  Masks in .png format are easy for anyone to open and visualize, but the final desired format before feeding the data into a segmentation network is .npy format (most common).  A .npy file is a NumPy matrix.

### Input image
#### Satellite image of trails

```python
image = Image.open('../data/sample_image.png')
print('Input image dimensions: {}'.format(image.size))
plt.figure(figsize = (8,8))
plt.imshow(image)
```

### Mask
#### Mask showing ROI (labeled trails)

Below is a mask in .png format.  The ROI is white (pixel value 255), and the background is black (pixel value 0).

```python
mask = Image.open('../data/sample_mask.png')
print('Mask dimensions: {}'.format(mask.size))
plt.figure(figsize = (8,8))
plt.imshow(mask)
```

Below is a mask in .npy format. The ROI is white (pixel value 1), and the background is black (pixel value 0).  The visualization for 1 and 0 are yellow and purple, respectively.

```python
mask_npy = np.load('../data/sample_mask.npy')
plt.figure(figsize = (8,8))
plt.imshow(mask_npy)
```

### How to transform .png image to NumPy (.npy) matrix

If you only have .png images for masks, you can transform them into NumPy matrices as follows.

First, create an array from the loaded mask (above), and divide the pixel values by 255 so that the ROI is denoted by 1s and the background is denoted by 0s.

```python
mask_matrix = np.array(mask)/255
print('Mask dimensions: {}'.format(mask_matrix.shape))
```

As you can tell from the mask dimensions above, there are 3 channels in the matrix for the RGB (red, green, blue) color channels.  You don't need all 3; you just need 1.  So next, extract just the first channel from the matrix.

```python
mask_matrix = mask_matrix[:,:,0]
print('Mask dimensions: {}'.format(mask_matrix.shape))
plt.figure(figsize = (8,8))
plt.imshow(mask_matrix)
```
